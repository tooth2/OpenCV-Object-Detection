```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
# Read in the image
image1 = cv2.imread('images/cars.jpg')
# Make a copy of the image
image_copy1 = np.copy(image1)
# Change color to RGB (from BGR)
image_copy1 = cv2.cvtColor(image_copy1, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy1)
```

```python
image2 = cv2.imread('images/pedestrian_bike.jpeg')
image_copy2 = np.copy(image2)
image_copy2 = cv2.cvtColor(image_copy2, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy2)
```

```python
# Convert to grayscale
gray1 = cv2.cvtColor(image_copy1, cv2.COLOR_RGB2GRAY)
gray1 = np.float32(gray1)
# Detect corners 
dst1 = cv2.cornerHarris(gray1, 2, 3, 0.04)
# Dilate corner image to enhance corner points
dst1 = cv2.dilate(dst1,None)
plt.imshow(dst1, cmap='gray')
```

```python
# Convert to grayscale
gray2 = cv2.cvtColor(image_copy2, cv2.COLOR_RGB2GRAY)
gray2 = np.float32(gray2)
# Detect corners 
dst2 = cv2.cornerHarris(gray2, 2, 3, 0.04)
# Dilate corner image to enhance corner points
dst2 = cv2.dilate(dst2,None)
plt.imshow(dst2, cmap='gray')
```

```python
# Define a threshold for extracting strong corners
# This value vary depending on the image and how many corners you want to detect
# Try changing this free parameter, 0.1, to be larger or smaller ans see what happens
thresh = 0.005*dst1.max()
# Create an image copy to draw corners on
corner_image1 = np.copy(image_copy1)
# Iterate through all the corners and draw them on the image (if they pass the threshold)
for j in range(0, dst1.shape[0]):
    for i in range(0, dst1.shape[1]):
        if(dst1[j,i] > thresh):
            # image, center pt, radius, color, thickness
            cv2.circle( corner_image1, (i, j), 1, (0,255,0), 1)
plt.imshow(corner_image1)
```

```python
thresh = 0.005*dst2.max()
# Create an image copy to draw corners on
corner_image2 = np.copy(image_copy2)
# Iterate through all the corners and draw them on the image (if they pass the threshold)
for j in range(0, dst2.shape[0]):
    for i in range(0, dst2.shape[1]):
        if(dst2[j,i] > thresh):
            # image, center pt, radius, color, thickness
            cv2.circle( corner_image2, (i, j), 1, (0,0,255), 1)
plt.imshow(corner_image2)
```
